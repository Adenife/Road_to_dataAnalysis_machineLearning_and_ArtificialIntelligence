```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
new = np.random.rand(12, 15)
sns.heatmap(new, annot = True)
```

```python
sns.heatmap(new, annot = True, vmax = 0, vmin = 1, cmap = 'summer')
```

```python
sns.heatmap(new, annot = True, vmax = 0, vmin = 1, cmap = 'RdBu')
```

```python
sns.heatmap(new, annot = True, vmax = 0, vmin = 1, cmap = 'RdBu', linewidth = 0.2)
```

```python
sns.heatmap(new, annot = True, vmax = 0, vmin = 1, cmap = '', linewidth = 0.2, fmt = 'f')
```
